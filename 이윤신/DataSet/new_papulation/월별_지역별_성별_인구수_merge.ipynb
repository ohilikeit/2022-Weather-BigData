# 일별데이터로 변환

```python
pretreatment_cpy
```

```python
pretreatment_2 = pd.DataFrame({
     "area":['이곳','저곳'],
     "sex":[1,2],
     "yymmdd":['3','4'],
     "tot_person":[5,6]
})
```

```python
def dd_insert(tot_date , i):
  global pretreatment_2
  for j in range(1,tot_date+1):
      date = ''
      if j<10:
        date = '0'+str(j)
      else:
        date = str(j)
      new_data_frame = pd.DataFrame({
          "area":list(pretreatment_cpy.iloc[i:i+2,0]),
          "sex":list(pretreatment_cpy.iloc[i:i+2,1]),
          "yymmdd":[str(pretreatment_cpy.iloc[i,2])+date,str(pretreatment_cpy.iloc[i,2])+date],
          "tot_person":list(pretreatment_cpy.iloc[i:i+2,3]),
      })
      pretreatment_2 = pd.concat([pretreatment_2,new_data_frame])
  

for i in range(0,len(pretreatment_cpy),2):
  month = int(pretreatment_cpy.iloc[i,2]) % 100
  if(month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or month == 10 or month == 12):
    dd_insert(31,i)
  elif (month == 4 or month == 6 or month == 9 or month == 11):
    dd_insert(30,i)
  else:
    year = int(int(pretreatment_cpy.iloc[i,2])/100)
    if(year == 2012 or year == 2016):
      dd_insert(29,i)
    else:
      dd_insert(28,i)
```
